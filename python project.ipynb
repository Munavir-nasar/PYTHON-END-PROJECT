```python
# 1


team_distribution = data['Team'].value_counts()


percentage_split = (team_distribution / len(data)) * 100


distribution_df = pd.DataFrame({
    'Count': team_distribution,
    'Percentage': percentage_split
})

print(distribution_df)
```

```python
import matplotlib.pyplot as plt


fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 8))


ax1.bar(distribution_df.index, distribution_df['Count'], color='skyblue')
ax1.set_title('Number of Players per Team', fontsize=14)
ax1.set_xlabel('Team', fontsize=12)
ax1.set_ylabel('Count', fontsize=12)
ax1.tick_params(axis='x', rotation=90)  


ax2.pie(
    distribution_df['Percentage'],
    labels=distribution_df.index,
    autopct='%1.1f%%',
    startangle=140,
    colors=plt.cm.Paired.colors 
ax2.set_title('Percentage of Players per Team', fontsize=14)


plt.tight_layout()
plt.show()
```

```python
# 2

position_groups = data.groupby('Position')
position_dict = {position: group for position, group in position_groups}

for position, group in position_dict.items():
    print(f"Position: {position}")
    print(group.head(), '\n')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


position_distribution = data['Position'].value_counts()


plt.figure(figsize = (8,6))
sns.barplot(x = position_distribution.index, y = position_distribution.values, palette = "viridis")
plt.title('Number of Players by Position', fontsize = 16)
plt.xlabel('Position', fontsize = 12)
plt.ylabel('Player Count', fontsize = 12)
plt.xticks(fontsize = 10)
plt.show()


plt.figure(figsize = (8,8))
plt.pie(position_distribution.values, labels = position_distribution.index, autopct = '%1.1f%%',
        startangle = 140, colors = sns.color_palette('coolwarm', len(position_distribution)))
plt.title('Percentage  Players by Position', fontsize = 16)
plt.show()
```

```python
# 3

import pandas as pd


bins = [0,20,25,30,35,40] #Age range
labels = ['<20','20-25','26-30','31-35','>=35']

data['Age Group'] = pd.cut(data['Age'], bins = bins, labels = labels, right = False)

age_group_distribution = data['Age Group'].value_counts().sort_index()
age_group_distribution.name = "Age Distribution"

predominant_age_group = age_group_distribution.idxmax()

print("Distribution of players by age group:")
print(age_group_distribution)
print("\nPredominant age group:", predominant_age_group)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns


plt.figure(figsize = (10,6))
sns.barplot(x = age_group_distribution.index, y = age_group_distribution.values, palette = "rocket")
plt.title("Distribution of Players by Age Group", fontsize = 12)
plt.xlabel('Age Group', fontsize = 12)
plt.ylabel('Player Count', fontsize = 12)
plt.xticks(fontsize = 10)
plt.show()

plt.figure(figsize = (8,8))
plt.pie(age_group_distribution.values, labels = age_group_distribution.index, autopct = '%1.1f%%',
        startangle = 140, colors = sns.color_palette('rocket', len(age_group_distribution)))
plt.title("Percentage of Players by Age Group", fontsize = 12)
plt.show()
```

```python
# 4

team_salary_expenditure = data.groupby('Team')['Salary'].sum().sort_values(ascending = False)
team_salary_expenditure.name = "Team Salary"


position_salary_expenditure = data.groupby('Position')['Salary'].sum().sort_values(ascending = False)
position_salary_expenditure.name = "Position Salary"


highest_team_salary = team_salary_expenditure.idxmax()
highest_position_salary = position_salary_expenditure.idxmax()

print("Total Salary Expenditure by Team:")
print(team_salary_expenditure, "\n")
print(f"Team with the highest salary expenditure: {highest_team_salary} (${team_salary_expenditure.max():,.2f})\n")

print("Total Salary Expenditure by Position:")
print(position_salary_expenditure, "\n")
print(f"Position with the highest salary expenditure: {highest_position_salary} (${position_salary_expenditure.max():,.2f})\n")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize = (12, 6))
sns.barplot(x = team_salary_expenditure.index, y = team_salary_expenditure.values, palette = "inferno")
plt.title('Total Salary Expenditure by Team', fontsize = 16)
plt.xlabel('Team', fontsize = 12)
plt.ylabel('Total Salary (in $)', fontsize = 12)
plt.xticks(rotation = 45, ha = 'right', fontsize = 10)
plt.tight_layout()
plt.show()

plt.figure(figsize = (10, 6))
sns.barplot(x = position_salary_expenditure.index, y = position_salary_expenditure.values, palette = "Set2")
plt.title('Total Salary Expenditure by Position', fontsize = 16)
plt.xlabel('Position', fontsize = 12)
plt.ylabel('Total Salary (in $)', fontsize = 12)
plt.xticks(fontsize = 10)
plt.tight_layout()
plt.show()
```

```python
# 5


correlation = data['Age'].corr(data['Salary'])
print(f"The correlation between Age and Salary is: {correlation:.2f}")

if correlation > 0:
    correlation_type = 'Positive Correlation'
elif correlation < 0:
    correlation_type = 'Negative Correlation'
else:
    correlation_type = 'No Correlation'

print(f"The correlation between Age and Salary is: {correlation_type}")
```

```python
correlation_matrix = data[['Age','Salary']].corr()

plt.figure(figsize = (6,4))
sns.heatmap(correlation_matrix, annot = True, cmap = "magma", fmt = '.2f', linewidth = 0.5, cbar = True, square = True)
        
plt.title("Correlation Between Age & Salary", fontsize = 16)
plt.xticks(fontsize = 10)
plt.yticks(fontsize = 10)
plt.tight_layout()
plt.show()
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4))
sns.regplot(x='Age', y='Salary', data=data, scatter_kws={'alpha': 0.6}, line_kws={'color': 'black'})

plt.title("Scatter Plot with Regression Line: Age vs Salary", fontsize=10)
plt.xlabel("Age", fontsize=8)
plt.ylabel("Salary", fontsize=8)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.tight_layout()
plt.show()
```
